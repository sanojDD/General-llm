```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    template="""
    You are a helpful AI chatbot assistant. You must answer **strictly** based on the following transcript context.

    If the answer is not directly found in the transcript, just respond:
    "It's not discussed in the provided context."

    --------------------
    Transcript:
    {context}
    --------------------

    Question: {question}

    Answer (based only on the transcript above):
    """,
    input_variables=["context", "question"]
)
```

```python
question = "What is the qualification of the user?"
```

```python
retrieved_docs =  retrievers.invoke(question)
```

```python
retrieved_docs
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
print(context_text)
```

```python
final_prompt = template.invoke({"context": context_text, "question": question})
print(final_prompt)
```

```python
answer = groq_llm.invoke(final_prompt)
print(answer.content)
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

rag_chain = (
    {"context": retrievers | format_docs, "question": RunnablePassthrough()}
    | template
    | groq_llm
    | StrOutputParser()
)
```

```python
result2 = rag_chain.invoke("what is the qualification of the user?")
print(result2)
```

```python
result2 = rag_chain.invoke("Is theh user suitable fopr AI ML jobs or intern ?")
print(result2)
```
